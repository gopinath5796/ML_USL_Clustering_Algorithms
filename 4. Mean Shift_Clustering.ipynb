```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
bandwidth = estimate_bandwidth(X, quantile=0.1, n_samples=100)
y_MS = MeanShift(bandwidth=bandwidth, bin_seeding=True)
y_Meanshift = y_MS.fit_predict(X)
print(y_Meanshift)
```

```python
cluster_centers_ = y_MS.cluster_centers_
print(cluster_centers_)
```

```python
n_clusters_ = len(cluster_centers_)
print("number of estimated clusters : %d" % n_clusters_)
```

```python
supervised=pd.DataFrame(dataset)
supervised
```

```python
supervised['Cluster_group']=y_Meanshift
supervised
```

```python
supervised.to_csv("Mean Shift_cluster.csv",index=False)
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```
